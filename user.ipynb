```python
from threeDVec import *

# User login/register function:
# 1) If the username exists, enter PIN verification (up to 3 attempts)
# 2) If the username does not exist, append new username and PIN
# Return value: 1 means enter main menu, 0 means login failed and terminate

def user():
    try:
        # Try to read the user file
        users = open('user.txt', 'r')
    except FileNotFoundError:
        # Create the file if missing, and keep it readable for readlines() below
        users = open('user.txt', 'a+')
        users.seek(0)

    login_or_register = input('Please input your username',)
    user_lines = users.readlines()
    users.close()

    # Registered user: read and verify PIN
    if (f'{login_or_register}\n') in user_lines:
        users = open('user.txt', 'r')
        password = input('Username found. Please enter your PIN:',)

        # Find the PIN on the line immediately after this username
        while True:
            line = users.readline()
            if ((f'{login_or_register}\n') == line):
                truepassword = users.readline().strip()
                break
            else:
                continue

        n = 0
        # Allow at most 3 PIN attempts
        while True:
            if (password == truepassword):
                users.close()
                return 1
            else:
                if n == 2:
                    users.close()
                    print('Login failed. Program terminating due to 3 consecutive incorrect trials.')
                    return 0
                else:
                    password = input(f'Incorrect PIN. You have {2-n} trials remaining',)
                    n += 1
                    continue
    else:
        # New user registration: write username and PIN
        users = open('user.txt', 'a')
        users.write(f'{login_or_register}\n')
        password = input('New user detected. Please set your PIN:',)
        users.write(f'{password}\n')
        users.close()
        return 1
def run_main_menu(m):
    # Main menu loop: enter feature menu after successful login
    while True:
        if m == 1:
            print('=' * 30)
            print(f"{'3D VECTOR MAIN MENU':^30}")
            print('=' * 30)
            print('1. Enter 3D Vector Components (a, b, c)')
            print('2. Calculate Azimuthal Angle (Distinction)')
            print('3. Exit Program')
            print('=' * 30)
            print('Please select an option (1-3):')

            # Menu input: integer input is assumed here
            x = int(input())

            if x == 1:
                # Feature 1: read two vectors and compare their squared magnitudes
                print('Enter the components of the 3D vector:')

                while True:
                    a1 = float(input('please enter a:',))
                    if a1 == 0:
                        print('a cannot be zero, please input again')
                        continue
                    else:
                        break
                b1 = float(input('please enter b:',))
                c1 = float(input('please enter c:',))
                vector1 = threeDVec(a1, b1, c1)

                while True:
                    a2 = float(input('please enter a:',))
                    if a2 == 0:
                        print('a cannot be zero, please input again')
                        continue
                    else:
                        break
                b2 = float(input('please enter b:',))
                c2 = float(input('please enter c:',))
                vector2 = threeDVec(a2, b2, c2)

                # compare_magnitude_squared takes the other vector's squared magnitude
                t = vector1.compare_magnitude_squared(vector2.magnitude_squared())
                print(f'The result is {t}')
                continue

            elif x == 2:
                # Feature 2: calculate azimuthal angle (Distinction)
                print('Enter the components of the 3D vector:')

                while True:
                    a = float(input('please enter a:',))
                    if a == 0:
                        print('a cannot be zero, please input again')
                        continue
                    else:
                        break
                b = float(input('please enter b:',))
                c = float(input('please enter c:',))

                # Read Taylor-series term count terms (must be an integer)
                while True:
                    try:
                        terms = int(input('Please input terms',))
                        break
                    except ValueError:
                        print("terms must be an integer, please input again")

                vector = vazmAng(a, b, c)
                angle = vector.get_azimuthal(terms)
                print(f'Azimuthal angle:{angle} rad')
                continue

            elif x == 3:
                print('Exiting program. Goodbye!')
                break
            else:
                print('Invalid option. Please select a valid option (1-3).')
                continue
        else:
            break

if __name__ == '__main__':
    m = user()
    run_main_menu(m)
```
